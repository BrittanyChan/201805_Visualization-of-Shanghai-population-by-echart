```python
import numpy as np
import seaborn as sns  
import matplotlib.pyplot as plt 

sns.set(style = "whitegrid")#设置样式  
font = {'family' : 'SimHei'}
plt.rc('font', **font)

x=df4.地区#X轴数据  
y=df4.出生人数#Y轴数据  
z=df4.已婚育龄妇女人数#用来调整各个点的大小s  
cm=plt.cm.get_cmap('RdYlBu')  
fig,ax = plt.subplots(figsize = (12,10))  
#注意s离散化的方法，因为需要通过点的大小来直观感受其所表示的数值大小  
#我所使用的是当前点的数值减去集合中的最小值后+0.1再*1000  
#参数是X轴数据、Y轴数据、各个点的大小、各个点的颜色  
bubble=ax.scatter(x,y,s=(z-np.min(z)+0.1)*1000,c=z,cmap=cm,linewidth=0.5,alpha=0.5)  
ax.grid()  
fig.colorbar(bubble)  
ax.set_xlabel('location',fontsize=15)#X轴标签  
ax.set_ylabel('newborn',fontsize=15)#Y轴标签
plt.title('各区户籍人口生育基本情况（2016）')
plt.savefig('各区户籍人口生育基本情况（2016）.png', dpi=100, bbox_inches='tight')
plt.show()  
```

```python
#-*- coding: utf-8 -*-
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.basemap import Basemap 
from matplotlib.patches import Polygon

font = {
    'family' : 'SimHei'
};
plt.rc('font', **font);

fig = plt.figure()
ax1 = fig.add_axes([0.1,0.1,0.8,0.8])
bmap = Basemap(projection='stere',lat_0=30.5,lon_0=121,
            llcrnrlon=120.8575366916139,llcrnrlat=30.67070630728205,
               urcrnrlon=121.9775260777007,urcrnrlat=31.49681439748955,
            rsphere=6371200.,resolution='l',area_thresh=10000,ax=ax1)
shp_info = bmap.readshapefile('gadm36_CHN_shp/gadm36_CHN_1','states')

for info, shp in zip(bmap.states_info, bmap.states):
    proid = info['NAME_1']
    if proid == 'Shanghai':
        poly = Polygon(shp,facecolor='w',edgecolor='b', lw=0.2)
        ax1.add_patch(poly)

bmap.drawmapboundary()   # 绘制边界
#map.fillcontinents()   # 填充大陆，发现填充之后无法显示散点图，应该是被覆盖了
bmap.drawstates()        # 绘制州
bmap.drawcoastlines()    # 绘制海岸线
bmap.drawcountries()     # 绘制国家
bmap.drawcounties()      # 绘制县

bmap.drawparallels(np.arange(29,32,2),labels=[1,0,0,0])
bmap.drawmeridians(np.arange(119,122,2),labels=[0,0,0,1])
plt.title('Shanghai Province')
plt.savefig('fig_province3.png', dpi=1000, bbox_inches='tight')
plt.clf()
plt.close()
```
